```python
dados = {
    "lingua" : ["ingles", "outros"],
"total": [total_de_ingles, total_do_resto]
}
```

```python
pd.DataFrame(dados)#criar um dataframe nesses dados
```

```python
sns.barplot(x="lingua", y="total", data = dados)
```

```python
#agr temos um grafico comparando ingles com os outros - agr tambem poderemos ver como estão as linguas dentro desses outros
```

```python
plt.pie(dados["total"], labels=dados["lingua"])
plt.show()
```

#Revisando o papel da média, mediana, medidas de tendência central dispersão, desvio padrão, box plot, histograma.

```python
import pandas as pd

notas = pd.read_csv("ratings.csv")
notas.columns = ["usuarioId", "filmeId", "nota", "momento"]
filmes = pd.read_csv("movies.csv")
```

```python
notas.head(2)
```

```python
#quero saber a media da nota de cada filme
```

```python
notas_do_toy_story = notas.query("filmeId==1")
notas_do_jumanji = notas.query("filmeId==2")
print(len(notas_do_toy_story), len(notas_do_jumanji))
```

```python
print("Nota média do Toy Story: %.2f" % notas_do_toy_story.nota.mean()) #media
print("Nota média do jumanji: %.2f" % notas_do_jumanji.nota.mean())
```

```python
print("Nota média do Toy Story: %.2f" % notas_do_toy_story.nota.median())#mediana
print("Nota média do jumanji: %.2f" % notas_do_jumanji.nota.median())
```

```python
#media e mediana tendem a resumir nossos valores em um unico dado, perdendo a informação de como os dados se distribuem
#Numpy = trabalhar com arrays(series) numericas
```

```python
import numpy as np
filme1 = np.append(np.array([2.5]*10),np.array([3.5]*10))
filme2 = np.append(np.array([5]*10), np.array([1]*10))
```

```python
filme1.mean()
```

```python
filme2.mean()
```

```python
#perdemos a distribuição dessas notas
```

```python
np.median(filme1)
```

```python
np.median(filme2)
```

```python
#a media e a mediana são iguais, mas sabemos que elas são bem diferentes, entao usar um numero pontual não faz sentido pois queremos analisar a situação como um todo
```

```python
sns.distplot(filme1)
```

```python
plt.hist(filme1)
plt.hist(filme2)
```

```python
plt.boxplot([filme1,filme2])
```

```python
sns.boxplot(notas_do_toy_story.nota)
sns.boxplot(notas_do_jumanji.nota)
```

```python
plt.boxplot([notas_do_toy_story.nota,notas_do_jumanji.nota])
```

```python
sns.boxplot(x = "filmeId", y ="nota", data = notas.query("filmeId in[1,2,3,4,5]"))
```

```python
#desvio padrão (standard deviation) diz o quanto os dados fogem de uma tendencia central
```

```python
notas_do_toy_story.nota.std()
```

```python
notas_do_jumanji.nota.std()
```

```python
print(np.std(filme1), np.std(filme2))
```
